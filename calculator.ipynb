```python
from tkinter import*
r=Tk()
r.geometry("570x620+300+200")
r.configure(bg="#17161b")
r.resizable(0,0)
r.title("Calculator")
Equation=""
def show(value):
    global Equation
    Equation+=value
    Label_result.config(text=Equation)

def clear():
    global Equation
    Equation=""
    Label_result.config(text=Equation)
    
def Calculate():
    global Equation
    result=eval(Equation)
    Label_result.config(text=result)


main=Label(r,text=" ",width=25,height=2,font=("arial",30,))
main.place(x=0,y=2)
Label_result=Label(r,width=25,height=2,font=("arial",30))
Label_result.pack()








#row1
buttonc=Button(r,text="c",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:clear())
buttonc.place(x=10,y=110)
buttonper=Button(r,text="-",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("-"))
buttonper.place(x=150,y=110)
buttonm=Button(r,text="*",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("*"))
buttonm.place(x=290,y=110)
buttondi=Button(r,text="/",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("/"))
buttondi.place(x=430,y=110)

#row2
button9=Button(r,text="9",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("9"))
button9.place(x=10,y=220)
button8=Button(r,text="8",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("8"))
button8.place(x=150,y=220)
button7=Button(r,text="7",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("7"))
button7.place(x=290,y=220)
buttonplus=Button(r,text="+",width=5,height=3,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("+"))
buttonplus.place(x=435,y=220)

#row3
button6=Button(r,text="6",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("6"))
button6.place(x=10,y=320)
button5=Button(r,text="5",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("5"))
button5.place(x=150,y=320)
button4=Button(r,text="4",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("4"))
button4.place(x=290,y=320)

#row4
button3=Button(r,text="3",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("3"))
button3.place(x=10,y=420)
button2=Button(r,text="2",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("2"))
button2.place(x=150,y=420)
button1=Button(r,text="1",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("1"))
button1.place(x=290,y=420)
buttoneq=Button(r,text="=",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:Calculate())
buttoneq.place(x=430,y=420)

button0=Button(r,text="0",width=11,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("0"))
button0.place(x=10,y=520)

buttond=Button(r,text=".",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("."))
buttond.place(x=430,y=520)
buttonper=Button(r,text="%",width=5,height=1,fg="#fff",bg="#36c7f5",font=("arial",30,"bold"),command=lambda:show("%"))
buttonper.place(x=290,y=520)


mainloop()
```
